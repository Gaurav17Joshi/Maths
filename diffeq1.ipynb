```python
import autograd.numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0,2*np.pi,100)
import seaborn as sns
sns.set()

def sinn(x,b):
    ans = 0
    for i in range(1,b+1):
        if i%2 == 0:
            ans = ans + (-1)*np.sin(i*x)/i
        else:
            ans = ans + np.sin(i*x)/i
    return ans

y = [2*(sinn(t,10)) for t in x]

fig, ax = plt.subplots()
ax.plot(x,y,'-')
zeros = np.zeros(100)
ax.plot(x,zeros, 'k')
ax.plot(zeros,y,'k')

# Make a slider for demonstrating the funtion:
```

```python
# make animation for spinning sin finction
# fourier series for step function:
x = np.linspace(-1*np.pi,1*np.pi,200)

def stepp(x,b):
    ans = 0
    for i in range(1,2*b+1,2):
        ans = ans + np.sin(i*x)/i
    return ans

y = [0*np.pi + 2*(stepp(t,50)) for t in x]

fig, ax = plt.subplots()
ax.plot(x,y,'-')
zeros = np.zeros(200)
ax.plot(x,zeros, 'k')
ax.plot(zeros,y,'k')
```

```python
x = np.linspace(-1*np.pi,1*np.pi,200)

def sinh(x):
    return (np.exp(x)-np.exp(-x))

def xzeros(x,b):
    a = sinh(np.pi)
    ans = 1/(2*np.pi)
    # for i in range(1,b,2):
    #     ans = ans + - 2/np.pi*(np.cos(i*x)/i**2)
    for i in range(1,b):
        if (i%2 == 1):
            ans = ans - np.cos(i*x)/(np.pi*(i*i + 1)) + i*np.sin(i*x)/(np.pi*(i*i + 1))
        else:
            ans = ans + np.cos(i*x)/(np.pi*(i*i + 1)) - i*np.sin(i*x)/(np.pi*(i*i + 1))
    return ans*a

y = [xzeros(t,50) for t in x]
yr = [np.exp(t) for t in x]

fig, ax = plt.subplots(1,2)
fig.set(figwidth = 15, figheight= 10)
ax[0].plot(x,y,'-')
ax[1].plot(x,yr,'g')
zeros = np.zeros(200)
ax[0].plot(x,zeros, 'k')
ax[0].plot(zeros,y,'k')
ax[1].plot(x,zeros, 'k')
ax[1].plot(zeros,y,'k')
```
