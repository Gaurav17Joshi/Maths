```python
import autograd.numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0,2*np.pi,100)
import seaborn as sns
sns.set()

def sinn(x,b):
    ans = 0
    for i in range(1,b+1):
        if i%2 == 0:
            ans = ans + (-1)*np.sin(i*x)/i
        else:
            ans = ans + np.sin(i*x)/i
    return ans

y = [2*(sinn(t,10)) for t in x]

fig, ax = plt.subplots()
ax.plot(x,y,'-')
zeros = np.zeros(100)
ax.plot(x,zeros, 'k')
ax.plot(zeros,y,'k')

# Make a slider for demonstrating the funtion:
```
